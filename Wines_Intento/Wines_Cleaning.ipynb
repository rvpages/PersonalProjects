```python
import pandas as pd
```

```python
df = pd.read_csv('wines_vivino.csv')
df.head()
```

## Data Cleaning

#### Year Column

```python
#Create a column with the year
df['Year'] = df['Grape'].apply(lambda x: x.split()[-1])
```

```python
#Drop the year in Grape column
df['Grape'] = df['Grape'].apply(lambda x: x.replace(x.split()[-1],''))
```

```python
#Lets convert the column year into type: int64
df['Year'] = pd.to_numeric(df['Year'])
df['Year'].dtype
```

#### Price Column

```python
#Drop the $ in Price column
df['Price'] = df['Price'].apply(lambda x: x.replace('$',''))
```

```python
df['Price'] = pd.to_numeric(df['Price'])
df['Price'].dtype
```

```python
df.head()
```

#### Grape column

```python
def grape_simplifier(grape):
    if 'malbec' in grape.lower():
        return 'malbec'
    elif 'cabernet sauvignon' in grape.lower():
        return 'cabernet sauvignon'
    elif 'cabernet franc' in grape.lower():
        return 'cabernet franc'
    elif 'blend' in grape.lower():
        return 'blend'
    elif 'bonarda' in grape.lower():
        return 'bonarda'
    else:
        return 'other'
```

```python
df['Grape'] = df['Grape'].apply(grape_simplifier)
df.head(30)
```

```python
df.to_csv('wine_data_cleaned.csv',index = False)
```
